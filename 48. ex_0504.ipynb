```python
import requests as rq
import bs4

# 1. Lấy dữ liệu từ trang web
res = rq.get("https://www.pvl.vn/About")
res.raise_for_status()  # Kiểm tra lỗi HTTP

# 2. Phân tích HTML bằng BeautifulSoup
soup = bs4.BeautifulSoup(res.text, 'html.parser')
```

```python
# 3. Lấy tất cả văn bản từ trang web
def get_all_text(soup):
    # Loại bỏ các thẻ script và style
    for script in soup(["script", "style", "nav", "footer", "head", "meta"]):
        script.decompose()
    
    # Lấy text và làm sạch
    text = soup.get_text(separator='\n', strip=True)
    
    # Xóa các dòng trống
    lines = [line for line in text.split('\n') if line.strip()]
    return '\n'.join(lines)
```

```python

all_text = get_all_text(soup)

# 4. In kết quả
print("TOÀN BỘ NỘI DUNG VĂN BẢN TỪ TRANG WEB:")
print(all_text)
```
